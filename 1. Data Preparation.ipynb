```python
import pandas as pd
import numpy as np
```

```python
review = pd.read_csv('review.csv')
review
```

```python
business = pd.read_csv('yelp_academic_dataset_business.csv')
business
```

```python
all_data = pd.merge(review, business, on='business_id')
all_data
```

```python
all_data_open = all_data[all_data['is_open']==1]
all_data_open
```

```python
all_data_open_res = all_data_open[all_data_open['categories'].str.contains('Restaurant', na=False)]
all_data_open_res
```

```python
user = pd.read_csv('yelp_user.csv')
```

```python
all_data_open_res = pd.merge(all_data_open_res, user, on='user_id')
all_data_open_res
```

```python
all_data_open_res.info()
```

```python
all_data_open_res = all_data_open_res.rename({'stars_x':'stars_r', 'useful_x':'useful_r', 'funny_x':'funny_r', 'cool_x':'cool_r',
                  'name_x': 'name_b', 'stars_y':'stars_b', 'review_count_x':'review_count_b',
                  'funny_y':'funny_u', 'review_count_y':'review_count_u', 'useful_y':'useful_u', 'name_y':'name_u', 'cool_y':'cool_u'}, axis=1)
```
